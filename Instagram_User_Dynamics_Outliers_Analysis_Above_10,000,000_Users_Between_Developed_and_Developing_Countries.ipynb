Removing rows with missing values can be the right choice in certain situations. Here are the reasons why removing data with missing values is appropriate in this context:

1. Preventing Bias in Analysis

 - If missing data is not randomly distributed, for example, if the data is missing for specific reasons that could affect the analysis, removing rows with missing values can help prevent bias. For instance, if Instagram user data in certain countries is missing due to data collection issues in those countries, the missing data may not represent the entire population, and removing it would help maintain the accuracy of the analysis.

2. Data Size Remains Sufficiently Large

 - The dataset has 226 rows, with only 21 missing values (about 9.29%). Even with some data loss, the remaining data (205 rows) is still large enough for analysis. Therefore, removing rows with missing values may not significantly impact the final results of the analysis.

3. Minimizing the Risk of Errors from Imputation

 - Filling in missing values (imputation) can introduce errors, especially if the imputation method used is not entirely accurate. For example, if missing values are filled with the mean or median, it could lead to the loss of important data variation or introduce values that do not accurately reflect the real situation. By removing rows with missing values, the analysis is based only on complete data, reducing the risk of errors caused by inaccurate imputation.

4. Decision Based on Analytical Needs

 - If only descriptive analysis or data exploration is being performed, removing rows with missing values is a simpler and safer approach because it focuses on complete data without the need to fill in or process missing values. This reduces complexity without losing much information, as the remaining data is still fairly representative.

5. Data Cleaning for Predictive Models

 - If planning to use this data to build predictive models, many machine learning algorithms require complete datasets with no missing values. Removing rows with missing values is a straightforward solution to ensure the model can be trained without issues.

### Initial Statistical Test

Initial statistical tests were conducted to check the normality of the distribution of numerical data using the **Kolmogorov-Smirnov**, **Lilliefors**, **Shapiro-Wilk**, and **D'Agustino-Pearson tests**.

#### Kolmogorov Smirnov

```python
from scipy.stats import kstest

for columns in numeric.columns:
    stat, pval = kstest(data[columns],'norm')
    if pval < 0.05:
        print(f'Distribution {columns} is not normal')
    else:
        print(f'Distribution {columns} is normal')
```

#### Liliefors Test

```python
from statsmodels.stats.diagnostic import lilliefors

for columns in numeric.columns:
    stat, pval = lilliefors(data[columns])
    if pval < 0.05:
        print(f'Distribution {columns} is not normal')
    else:
        print(f'Distribution {columns} is normal')
```

#### Shapiro Wilk

```python
from scipy.stats import shapiro

for columns in numeric.columns:
    stat, pval = shapiro(data[columns])
    if pval < 0.05:
        print(f'Distribution {columns} is not normal')
    else:
        print(f'Distribution {columns} is normal')
```

#### D'Agustino Pearson

```python
from scipy.stats import normaltest

for columns in numeric.columns:
    stat, pval = normaltest(data[columns])
    if pval < 0.05:
        print(f'Distribution {columns} is not normal')
    else:
        print(f'Distribution {columns} is normal')
```

The results of the **Kolmogorov-Smirnov**, **Lilliefors**, **Shapiro-Wilk**, and **D'Agustino-Pearson tests** show that all numeric columns in the Instagram data are **not normally distributed**.

### IQR (INTER QUARTILE RANGE) * 1.5

```python
numeric = data.select_dtypes(include = 'number')
list_of_outliers = []
for columns in numeric.columns:
    Q3 = numeric.describe().loc['75%',columns]
    Q1 = numeric.describe().loc['25%',columns]
    iqr = Q3 - Q1
    whis = 1.5 * iqr
    upper_limit = Q3 + whis
    lower_limit = Q1 - whis

    amount_of_outliers = len(numeric[(numeric[columns] > upper_limit) | (numeric[columns] < lower_limit)])
    list_of_outliers.append(amount_of_outliers)

pd.DataFrame({
    'Column Name': numeric.columns,
    'Number of Outliers': list_of_outliers
})
```

```python
def classify_country_status(country_name):
    developed_countries = [
        "United States", "Germany", "Canada", "Australia", "United Kingdom", "Japan", "South Korea",
        "France", "Italy", "Spain", "Poland", "Taiwan"
    ]
    if country_name in developed_countries:
        return "Developed Countries"
    else:
        return "Developing Countries"

result_users_2024['Economic_Status'] = result_users_2024['country'].apply(classify_country_status)
result_users_2024.reset_index(drop = True, inplace = True)
result_users_2024
```

# INFERENTIAL STATISTICS

```python
import pandas as pd
from scipy.stats import mannwhitneyu
from statsmodels.stats.proportion import proportions_ztest
```

## DEVELOPING COUNTRIES USERS VS DEVELOPED COUNTRIES USERS

```python
status_counts = result_users_2024['Economic_Status'].value_counts()
status_counts
```

```python
plt.figure(figsize=(8, 6))
plt.bar(status_counts.index, status_counts.values, color=['red','green'])

plt.xlabel('Economic Status')
plt.ylabel('Number of Countries')
plt.title('Number of Countries by Economic Status')

plt.show()
```

```python
developing_count = len(result_users_2024[result_users_2024['Economic_Status'] == 'Developing Countries'])
developing_count
```

```python
developed_count = len(result_users_2024[result_users_2024['Economic_Status'] == 'Developed Countries'])
developed_count
```

```python
# Calculate the total number of countries
total_count = len(result_users_2024)
total_count
```

```python
# Expected proportion for 'Developing Countries' (based on this data)
expected_prop_developing = developing_count / total_count
expected_prop_developing
```

```python
# Hypothesis
H0 = f"The proportion of countries with the economic status 'Developing Countries' is {expected_prop_developing:.3f}."
Ha = "The proportion of countries with the economic status 'Developing Countries' is not equal to the expected proportion."

# Z-test of proportions for economic status 'Developing Countries'
stat, pval = proportions_ztest(
                    developing_count,
                    total_count,
                    expected_prop_developing
)

print(f'Z Statistics: {stat}')
print(f'P-Value: {pval}')

# Determine whether H0 is accepted or rejected
if pval < 0.05:
    print(f'H0 is rejected so {Ha}')
else:
    print(f'H0 is accepted so that {H0}')
```

Since H0 is accepted, it can be concluded that the observed proportion of “Developing Countries” (21 out of a total of 33 countries, i.e. 21/33 ≈ 0.636) is not significantly different from the hypothesized proportion (0.636).

## MEDIAN OF DEVELOPING COUNTRIES USERS VS DEVELOPED COUNTRIES USERS

```python
sum_increase_by_users = result_users_2024.groupby('Economic_Status')['increase_in_users'].sum()
sum_increase_by_users
```

```python
median_increase_by_users = result_users_2024.groupby('Economic_Status')['increase_in_users'].median()
median_increase_by_users
```

```python
plt.figure(figsize=(8, 6))
plt.bar(status_counts.index, median_increase_by_users.values, color=['red','green'])

plt.xlabel('Economic Status')
plt.ylabel('Median Increase Users of Countries')
plt.title('Number Median of Countries by Economic Status')

plt.show()
```

```python
from scipy.stats import mannwhitneyu

developed_increase = result_users_2024[result_users_2024['Economic_Status'] == 'Developed Countries']['increase_in_users']
developing_increase = result_users_2024[result_users_2024['Economic_Status'] == 'Developing Countries']['increase_in_users']

stat, pval = mannwhitneyu(developed_increase, developing_increase)

print("t-statistic:", stat)
print("p-value:", pval)

# Hypothesis:
# H0: There is no difference in the median increase in Instagram users between developed and developing countries.
# H1: There is a difference in the median increase in Instagram users between developed and developing countries.

if pval < 0.05:
    print("Reject H0, there is a significant difference.")
else:
    print("Failed to reject H0, there is no significant difference.")
```

The test results show that there is no significant difference between the median increase in users for "Developed Countries" and "Developing Countries". With a t-statistic value of 150.0 and a p-value of 0.379, the null hypothesis (H0) is not rejected, meaning the observed difference in the median increase in users between the two categories is not statistically significant enough.
